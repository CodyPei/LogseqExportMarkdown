```python
import os
```

```python
source_path = '/Users/peiel/Desktop/old_blog/'
blog_page_path = '/Users/peiel/Library/Mobile Documents/com~apple~CloudDocs/PKM/logseq/pages/blog.md'
logseq_page_dir_path = '/Users/peiel/Library/Mobile Documents/com~apple~CloudDocs/PKM/logseq/pages/'
```

```python
def get_categories(lines):
    idx = 0
    for line in lines:
        if line.strip() == 'categories:':
            return lines[idx + 1].replace('- ', '').strip()
        idx = idx + 1


def get_title(lines):
    for line in lines:
        if 'title:' in line:
            return line.replace('title:', '').strip()


def get_tags(lines):
    tags = []
    find = False
    for line in lines:
        if find and '-' in line:
            tag = line.replace('- ', '').strip()
            if ' ' in tag:
                tags.append('[[' + tag + ']]')
            else:
                tags.append(tag)
        else:
            find = False
        if 'tags' in line:
            find = True
    tags_str = ''
    for tag in tags:
        tags_str = tags_str + ' #' + tag
    if tags_str.strip != '':
        return ' ' + tags_str.strip()
    return tags_str.strip()


def get_created_time(lines):
    for line in lines:
        if 'date:' in line:
            return line.replace('date:', '').strip().replace('/', '-')


def find_categories_idx(bolg_lines, categories):
    idx = 0
    for line in bolg_lines:
        if ('- ## ' + categories) in line:
            return idx
        idx = idx + 1
    return idx
```

```python
def save_blog_page(lines):
    # 解析内容
    categories = get_categories(lines)
    title = get_title(lines)
    tags = get_tags(lines)
    created_time = get_created_time(lines)
    # 写入内容
    print(title, categories, tags, created_time)
    with open(blog_page_path) as f:
        blog_lines = f.readlines()

    idx = 0
    for line in blog_lines:
        if not '\n' in line:
            blog_lines[idx] = line + '\n'
        idx = idx + 1

    insert_idx = find_categories_idx(blog_lines, categories)
    if insert_idx == len(blog_lines):
        blog_lines.insert(insert_idx, '- ## ' + categories + '\n')
    insert_idx = insert_idx + 1
    blog_lines.insert(insert_idx, ('\t- [[' + title + ']] #' + (created_time + tags).strip() + '\n'))

    with open(blog_page_path, 'wt', encoding='utf-8') as f:
        f.write(''.join(blog_lines))
    return title

```

```python
def find_second_hyphen_idx(lines):
    times = 0
    idx = 0
    for line in lines:
        if '---' == line.strip():
            times = times + 1
        if times == 2:
            return idx
        idx = idx + 1
    return -1


def save_content(lines, title):
    # 删除头部内容
    second_hyphen_idx = find_second_hyphen_idx(lines)
    print(second_hyphen_idx)
    lines = lines[second_hyphen_idx + 1:]
    print(logseq_page_dir_path + title + '.md')
    with open(logseq_page_dir_path + title + '.md', 'wt', encoding='utf-8') as f:
        f.write(''.join(lines))
```

```python
def import_blog(path):
    # 读取文件
    with open(path) as f:
        lines = f.readlines()
    # 保存 blog 主表
    title = save_blog_page(lines)
    # 报错内容
    save_content(lines, title)



```

```python
for name in os.listdir(source_path):
    import_blog(source_path + name)
```
